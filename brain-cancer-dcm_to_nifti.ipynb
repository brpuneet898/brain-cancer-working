```python
src_root = '/kaggle/input/brain-cancer-dataset/2025_DCM'
dst_root = '/kaggle/working/2025_DCM'
```

```python
import os
import shutil
import subprocess
from pathlib import Path
```

```python
for patient_folder in os.listdir(src_root):
    src_patient_dir = os.path.join(src_root, patient_folder)


    if not os.path.isdir(src_patient_dir):
        continue


    for root, dirs, files in os.walk(src_patient_dir):
        dcm_files = [f for f in files if f.lower().endswith('.dcm')]

        if dcm_files:
            rel_path = os.path.relpath(root, src_root)
            dst_dir = os.path.join(dst_root, rel_path)
            os.makedirs(dst_dir, exist_ok=True)

            tmp_dir = '/kaggle/working/tmp_dcm'
            if os.path.exists(tmp_dir):
                shutil.rmtree(tmp_dir)
            os.makedirs(tmp_dir)

            for f in dcm_files:
                shutil.copy2(os.path.join(root, f), os.path.join(tmp_dir, f))

            subprocess.run([
                'dcm2niix',
                '-z', 'n',         # No compression
                '-o', dst_dir,     # Output path
                '-f', '%s',        # Keep original filename
                tmp_dir
            ])

            shutil.rmtree(tmp_dir)

print("✅ All patients processed. Converted .nii files are in /kaggle/working/2025_DCM/")
```

```python
import shutil
import os
from IPython.display import FileLink

folder_to_zip = "/kaggle/working/2025_DCM"
zip_file_path = "/kaggle/working/2025_DCM_Output.zip"

if os.path.exists(folder_to_zip):
    shutil.make_archive(zip_file_path.replace('.zip', ''), 'zip', folder_to_zip)
    print("Folder zipped successfully.")

    display(FileLink(zip_file_path))
else:
    print(f"Folder not found: {folder_to_zip}")
```
