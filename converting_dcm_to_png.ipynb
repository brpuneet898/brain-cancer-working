```python
import os
import pydicom
import matplotlib.pyplot as plt
```

```python
dcm_folder = '/content/'
```

```python
dcm_files = [f for f in os.listdir(dcm_folder) if f.endswith('.dcm')]
```

```python
first_file_path = os.path.join(dcm_folder, dcm_files[0])
ds = pydicom.dcmread(first_file_path)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(ds.pixel_array, cmap=plt.cm.gray)
plt.title(f"Patient ID: {ds.PatientID}")
plt.axis('off')
plt.show()
```
